```python
# Import pandas for data clean
import pandas as pd
```

```python
citi_bike_df = pd.read_csv('CitiBikeData.csv')
citi_bike_df
```

```python
#  Drop nan values
citi_bike_df = citi_bike_df.dropna(how='any').reset_index(drop=True)
```

```python
citi_bike_df.dtypes
```

```python
# replace the gender numbers
citi_bike_df = citi_bike_df.replace({'gender': {2: 'female', 1: 'male'}})
citi_bike_df = citi_bike_df[citi_bike_df.gender != 0]
citi_bike_df.head(20)
```

```python
# change the datatype for start and end times and add minutes column
citi_bike_df.dtypes
citi_bike_df['Dates'] = pd.to_datetime(citi_bike_df['starttime']).dt.date
citi_bike_df['Start Time'] = pd.to_datetime(citi_bike_df['starttime'])
citi_bike_df['Stop Time'] = pd.to_datetime(citi_bike_df['stoptime'])
citi_bike_df['Duration Minutes'] = citi_bike_df['tripduration']/60
```
